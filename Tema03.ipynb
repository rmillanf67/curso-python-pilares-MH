## Operadores de decisión


IF

```python
if condicion:
    consecuencia
```

```python
edad = 59
if edad >= 18:
    print("Eres mayor de edad")
```

```python
edad = int(input("Introduce tu edad: "))
if edad >= 18:
    print("Eres mayor de edad")
```

### else

```python
if condicion:
    consecuencia
else:
    otra_consecuencia
```

```python
edad = int(input("Introduce tu edad: "))
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
```

### elif

```python
edad = int(input("Introduce tu edad: "))
if edad > 40:
    print("Tiene más de 40 años")
elif edad >= 16:
    print("Tienes entre 16 y 40 años")
else:
    print("Tienes edad diferente a las anteriores")
```

### Operador ternario

```python
consecuencia_cierta if condicion else consecuencia_falsa
```

```python
edad = int(input("Introduce tu edad: "))
texto_mayor = "Eres mayor de edad"
texto_menor = "Eres menor de edad"

print(texto_mayor) if edad >=18 else print(texto_menor)
```

### Operadores anidados

```python
edad = 25
nombre = "Luis"

if edad >= 18:
    if nombre == "Christian":
        print("Eres Christian y eres mayor de edad")
    else:
        print("Eres mayor de edad pero no eres Christian")
```

Ejercicio: Operador Ternario
Número Positivo o Negativo

Solicita un número y determina mediante un operador ternario si es:

- Positivo
- Negativo


```python
saldo = int(input("Introduce tu saldo: "))
cantidad = int(input("Introduce la cantidad a retirar: "))

if saldo >= cantidad:
    if cantidad % 100 == 0:
        print("Retiro exitoso")
    else:
        print("La cantidad debe ser múltiplo de 100")
if saldo < cantidad:
    if saldo == 0:
        print("Cuenta sin fondos")
    else:
        print("Saldo insuficiente")
```
