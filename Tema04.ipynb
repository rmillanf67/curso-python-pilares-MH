Operadores de iteración

 * while

```python
i = 1
while i <= 10:
    num = int(input("ingresa un numero:   "))
    print("El numero ingresado es: {}",format(num))
    if num % 2 == 0:
        break
    i+=1
```

While ...else

```python
i = 10
print("Contando hacia atras")
while i>=0:
    print(i)
    i-=1
else:
    print("Termino el conteo")    
```

## Ejercicio 1:

Contar letra a
  
Solicita al usuario una palabra y cuenta cuantas letras "a" contiene utilizando un siclo for in.


```python
palabra = input("Escribe una palabra:  ")
contador_a = 0
for letra in palabra:
    if letra == "a":
        contador_a += 1 

print("La letra a aparece {} veces en la palabra {} ".format(contador_a, palabra))
    

```

Comando continue

```python
for i in range(1, 21):
    if i % 2 == 0:
        continue
    print(i)
    
```

```python
for i in range(0, 21, 5):
    if i % 2 == 5:
        continue
    print(i)
```

```python
for i in range(101):
    if i % 2 == 0 and i % 5 == 0:
        continue
    print(i)
```

Bucles anidados

```python
for i in range(1, 11):
    print("Tabla del {}".format(i))
    for j in range(1, 11):
        print("{} x {} = {}".format(i, j, i*j))
```

Ejercicio 3

    Triangulo numerico

```python
for fila in range(1,7):
    for numero in range(1,fila+1):
        print(numero, end="")
    print()    
```

```python
for Fila in range(1,6):
    for Asiento in range(1,9):
        print("Fila   {}    Asiento   {}".format(Fila,Asiento,Fila+Asiento))
```
