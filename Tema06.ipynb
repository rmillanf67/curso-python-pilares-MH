```python
nombres = []
dptos = []

usuarios_cant = int(input("Cuantos usuarios deseas ingresar:   "))

for i in range(usuarios_cant):
    nombre = input("Ingerese el nombre del usuario:   ")
    nombres.append(nombre)
    print(nombres)
    dpto = input("Ingrese el departamento del usuario:   ")
    dptos.append(depto)
    print(dptos)
    
    dicc = {}
    dicc["nombres"] = nombres
    dicc["departamentos"] = dptos
    print(dicc)
```
